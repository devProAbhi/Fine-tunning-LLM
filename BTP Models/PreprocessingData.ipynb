```python
# drop a column

data.drop(columns=["Recommendations"], axis=1, inplace=True)
data.head()
```

```python
# prompt: removing rows where null values in specific column
data_copy=data
data_copy = data_copy[~data_copy["Effective Safeguards"].isnull()]
data_copy.shape
```

```python
#make empty dataframe with similar header

import pandas as pd
new_data = pd.DataFrame(columns=data.columns)
new_data.head()
```

```python

j=0
# Iterate through rows of the DataFrame
for index, row in data.iterrows():
  if pd.notnull(row['Effective Safeguards']):
    # Start from the current row and move upward until a non-null value is found
    Safeguards=row['Effective Safeguards']
    for i in range(index, -1, -1):
      current_row = data.iloc[i]
      Node = current_row['Node']
      if pd.notnull(Node):
        break  # Exit the loop if a non-null value is found in the 'Node' column
    for i in range(index, -1, -1):
      current_row = data.iloc[i]
      Deviation = current_row['Deviation']
      if pd.notnull(Deviation):
        break
    for i in range(index, -1, -1):
      current_row = data.iloc[i]
      Cause = current_row['Cause']
      if pd.notnull(Cause):
        break
    for i in range(index, -1, -1):
      current_row = data.iloc[i]
      Consequence = current_row['Consequence']
      if pd.notnull(Consequence):
        break
    for i in range(index, -1, -1):
      current_row = data.iloc[i]
      Risk_S = current_row['HPL Risk Matrix-S']
      if pd.notnull(Risk_S):
        break
    for i in range(index, -1, -1):
      current_row = data.iloc[i]
      Risk_L = current_row['HPL Risk Matrix-L']
      if pd.notnull(Risk_L):
        break
    for i in range(index, -1, -1):
      current_row = data.iloc[i]
      Risk_RR = current_row['HPL Risk Matrix-RR']
      if pd.notnull(Risk_RR):
        break
    new_data.loc[j]=[Node,Deviation,Cause,Consequence,Risk_S,Risk_L,Risk_RR,Safeguards]
    j=j+1
```
